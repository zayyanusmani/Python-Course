# **Encapsulation**

Encapsulation in Python describes the concept of bundling data and methods within a single unit. So, for example, when you create a class, it means you are implementing encapsulation. A class is an example of encapsulation as it binds all the data members (instance variables) and methods into a single unit.

```python
#  _ protected
#  __ are for private

class Abc:
  def __init__(self):
    self.a = "a"
    self._b = "b"
    self.__c = "c"


  def p(self):
    print(self.a)
    print(self._b)
    print(self.__c)

  def _r(self):
    print("Hey there!")

  def __s(self):
    print("Blue")

  def g(self):
    return self.__c


ob = Abc()

# print(ob.a)
# print(ob._b)
ob.g()
ob.p()
ob._r()
# ob.__s()
```

```python
class Abc:
  def __init__(self):
    self.a = "a" #public
    self._b = "b" # protected
    self.__c = "c" #private

    self.temp = self.__c # you can acces private variables by storing them in public vars

class Efg(Abc):
  def fun(self):
    print(self.a)
    print(self._b)
    print(self.temp)
    print(self.__c)

ob = Efg()
ob.fun()
```
